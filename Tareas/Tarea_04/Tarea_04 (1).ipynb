```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

datos = {
    "Día": ["Miercoles 6 de diciembre", "Jueves 7 de diciembre", "Viernes 8 de diciembre", "Sábado 9 de diciembre", "Domingo 10 de diciembre", "Lunes 11 de diciembre", "Martes 12 de diciembre", "Miercoles 13 de diciembre"],
    "Tiempo en Instagram": ["2 horas 6 minutos", "1 hora 35 minutos", "1 hora 31 minutos", "1 hora", "1 hora 43 minutos", "2 horas 3 minutos", "1 hora 29 minutos", "49 minutos"],
    "Tiempo en Tiktok": ["3 horas 4 minutos", "3 horas 48 minutos", "2 horas 57 minutos", "3 horas 14 minutos", "2 horas 57 minutos", "2 horas 48 minutos", "1 hora 15 minutos", "6 minutos"],
    "Tiempo en Whatsapp": ["3 horas 34 minutos", "2 horas 47 minutos", "1 hora 9 minutos", "1 hora 41 minutos", "1 hora 31 minutos", "1 hora 46 minutos", "35 minutos", "23 minutos"],
    "Tiempo Total en pantalla": ["9 horas 53 minutos", "9 horas 40 minutos", "8 horas 11 minutos", "6 horas 41 minutos", "7 horas 53 minutos", "8 horas 17 minutos", "5 horas 7 minutos", "2 horas 21 minutos"]
}

df = pd.DataFrame(datos)

columnas_tiempo = ["Tiempo en Instagram", "Tiempo en Tiktok", "Tiempo en Whatsapp", "Tiempo Total en pantalla"]
df[columnas_tiempo] = df[columnas_tiempo].applymap(lambda x: sum(int(i) * (60 ** j) for j, i in enumerate(reversed(x.replace('horas', '').replace('hora', '').replace('minutos', '').split()))))

df.head()

plt.figure(figsize=(12, 6))
df.plot(x='Día', y=columnas_tiempo, kind='bar', stacked=True)
plt.xlabel("Día")
plt.ylabel("Tiempo (minutos)")
plt.title("Tiempo en Redes Sociales por Día")
plt.legend(title="Plataforma")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
